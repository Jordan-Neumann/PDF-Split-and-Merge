```python
import tkinter as tk
import PyPDF2
from PIL import Image, ImageTk
from tkinter.filedialog import askopenfile, askopenfilename

window = tk.Tk()

canvas = tk.Canvas(window, width = 500, height = 400)
canvas.grid(columnspan=5, rowspan = 5)

img = Image.open("logo.png")
img = ImageTk.PhotoImage(img)
img_label = tk.Label(image=img)
img_label.image = img
img_label.grid(column=3, row = 0)
        
INPUT_FILE = None
    
def select_input_file():
    global INPUT_FILE
    INPUT_FILE = askopenfile(parent=window, mode='rb',filetype=[("Pdf file", "*.pdf")])
    if not INPUT_FILE:
        raise ValueError('Cannot find file')

def create_output_file(input_file, page_start, page_end, new_file_name):

    if not page_start or not page_end:
        raise ValueError('Please input value')
    
    output_file = "{}.pdf".format(new_file_name)
    
    pdfReader = PyPDF2.PdfFileReader(input_file)
    pdfWriter = PyPDF2.PdfFileWriter()
    pdfOutputFile = open(output_file, 'wb')
    
    page_start -= 1
        
    for pageNum in range(page_start,page_end):
        pageObj = pdfReader.getPage(pageNum)
        pdfWriter.addPage(pageObj)

    pdfWriter.write(pdfOutputFile)
    pdfOutputFile.close() 

browse = tk.StringVar()
button = tk.Button(window, command = lambda:select_input_file(), textvariable = browse)
browse.set("Choose PDF")
button.grid(column=3,row=1)

page_start_label = tk.Label(window,text='Start: ')
page_start_label.grid(row=2,column=3)
page_start = tk.Entry(window,width=3)
page_start.grid(row=3,column=3)

page_end_label = tk.Label(window,text='End: ')
page_end_label.grid(row=4,column=3)
page_end = tk.Entry(window,width=3)
page_end.grid(row=5,column=3)

new_name_label = tk.Label(window,text='New Document Name: ')
new_name_label.grid(row=6,column=3)
new_name = tk.Entry(window,width=20)
new_name.grid(row=7,column=3)

browse2 = tk.StringVar()
button2 = tk.Button(window, command = lambda:create_output_file(INPUT_FILE, int(page_start.get()), int(page_end.get()), new_name.get()), textvariable = browse2)
browse2.set("Save PDF")
button2.grid(column=3,row=10)

window.mainloop()
```
